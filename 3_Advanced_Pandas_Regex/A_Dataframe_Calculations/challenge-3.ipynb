## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?
# Expectations: we can group by the customers and their sum of amount_spent
# Information is present inn orders, following code to try
# I am carrying the country as well, bc I need that for part two
customer_total_spending = orders.groupby(['CustomerID', 'Country'])['amount_spent'].sum().reset_index()
# worked 
display(customer_total_spending.head())
customer_total_spending.shape
```

```python
# Sub Problem 2 and 3
# Expectations: We need to get the quantiles first, then filter the customer_total_spending on them, add to table
# We know what quantiles we need (0.95 for vip, 0.75 for preferred)
min_spent = customer_total_spending['amount_spent'].min()
display(min_spent)
max_spent = customer_total_spending['amount_spent'].max()
display(max_spent)

quantiles = customer_total_spending['amount_spent'].quantile([0.75, 0.95])
customer_total_spending['category'] = pd.cut(
    customer_total_spending['amount_spent'],
    bins=[min_spent, quantiles[0.75], quantiles[0.95], max_spent],
    labels=['Regular','Preferred','VIP'])

# worked 
display(quantiles)
display(customer_total_spending.shape)
customer_total_spending
```

```python
# Subproblem 4: Show only VIP and Preferred customers 
vip_or_preferred_customers = customer_total_spending[customer_total_spending['category'].isin(['VIP', 'Preferred'])]

display(vip_or_preferred_customers['category'].unique())
vip_or_preferred_customers
```
