```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

1. Load the dataset into a notebook.

```python
data = pd.read_csv('./files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
display(data.shape)
data.head(1)
```

2. Create a barplot of the number of customers per Gender.

```python
gender_counts = data['Gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values)
```

3. Create a barplot of the number of customers per EmploymentStatus and Gender.

```python
e_g_data = data.pivot_table(index= ['Gender','EmploymentStatus'], values = ['Customer'], aggfunc = {'Customer':'count'}).reset_index()
sns.barplot(x='EmploymentStatus', y='Customer', hue='Gender', data=e_g_data)
```

4. Identify Measurements and Dimensions. Are they the same as both ones in your Pandas dataframe? Modify accordingly.

```python
def identify_dimensions_and_measures(data):
    dimensions = []
    measures = []
    for column in data.columns:
        if data[column].dtype == 'object':
            # If the data type is object (string) --> dimension
            dimensions.append(column)
        elif data[column].dtype in ['int64', 'float64']:
            # If the data type is int64 or float64 -->  measure
            measures.append(column)
    
    return dimensions, measures

dimensions, measures = identify_dimensions_and_measures(data)
print("Dimensions:", dimensions)
print("Measures:", measures)
```
